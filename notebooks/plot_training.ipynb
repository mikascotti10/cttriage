```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Leer CSV
csv_path = Path("../reports/training_log.csv")
df = pd.read_csv(csv_path)
print(df)

# Graficar
plt.figure(figsize=(8,5))
plt.plot(df["epoch"], df["train_loss"], label="Train Loss", marker="o")
plt.plot(df["epoch"], df["val_macro_f1"], label="Val Macro-F1", marker="o")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Evoluci√≥n del entrenamiento baseline")
plt.legend()
plt.grid(True)
plt.show()

```
