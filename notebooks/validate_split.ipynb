```python
from pathlib import Path
import pandas as pd
train = pd.read_parquet("../train_split.parquet")
val = pd.read_parquet("../val_split.parquet")

# Verificar que no hay estudios repetidos
overlap = set(train['image_path'].apply(lambda x: Path(x).parent.name)) & \
          set(val['image_path'].apply(lambda x: Path(x).parent.name))
print("Estudios en ambos conjuntos:", len(overlap))
```
